```python
from itertools import chain
import json
import re
import os
```

```python
def PIES(word):
    l = len(word)
    if l==1:
        return["S"]
    else:
        return ["B"]+["I"]*(l-2)+["E"]
```

```python
entity = ["n","nr","ns","nt","nx","nz","t"]
```

```python
def singlePreprocessing(sent):
    # 得到词
    sents = sent.strip("\n").split("\t")
    words =[i.split("/")[0] for i in sents]
    # 得到label
    label_pre = [i.split("/")[1] for i in sents]
    label = [[j]*len(i) for i,j in zip(words,label_pre)]
    ## 分词记录
    pos = [PIES(i) for i in words]
    pos = list(chain.from_iterable(pos))
    # 转成一维
    words = "".join(words)
    label = list(chain.from_iterable(label))
    label_end = []
    for i,lb in enumerate(label):
        if lb in entity:
            label_end.append(pos[i]+"-"+lb)
        else:
            label_end.append("O")
    
    return [words, label_end]
```

```python
def preprocessing(datafile, Vocab=None,Label=None):
    # 从文件中读取一行数据处理成[words, label]
    data =[]
    with open(datafile,encoding="utf-8") as f:
        for l in f:
            if not re.match(".*\[.*\].*",l):
                data.append(singlePreprocessing(l))
    filepath = os.path.dirname(datafile)
    
    # 生成vacab
    if Vocab == None:
        words = [i[0] for i in data]
        Vocab = list(set(chain.from_iterable(words)))
        Vocab.insert(0,"[PAD]")
        Vocab.append("[UNK]")
        with open(os.path.join(filepath,"vocab.json"),"w") as f:
            json.dump(Vocab,f)
    
    # 生成Label
    if Label == None:
        labels = [i[1] for i in data]
        Label = list(set(chain.from_iterable(labels)))
        with open(os.path.join(filepath,"label.json"),"w") as f:
            json.dump(Label,f)
            
    # 将data保存至文件
    with open(os.path.join(filepath,"wdata.json"),"w") as f:
        json.dump(data,f)
    return data
```
